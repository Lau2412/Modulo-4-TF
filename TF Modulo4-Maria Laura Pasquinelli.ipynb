```
import pandas as pd
import numpy as np
import pandas_datareader as dtr
import plotly.express as px
import plotly.graph_objects as go
import altair as alt
```

# **Obtener datos**

```
facturas= "https://raw.githubusercontent.com/cristiandarioortegayubro/BA/main/Datasets/Analisis_de_Facturas_de_Venta.csv"
```

# **Crear un DataFrame**

```
df = pd.read_csv(facturas)
df
```

```
df.info()
```

```
df["FECHA"]= pd.to_datetime(df.FECHA, infer_datetime_format=True)
df = df.sort_values(by="FECHA", ascending=True)
df = df.reset_index(drop=True)
df
```

# **Trabajando con el DataFrame**

## **Eliminando datos no disponibles y duplicados**

```
df.drop_duplicates()
df.dropna(how='any', inplace=True)
```

## **Insertando columnas**

```
df.insert(8, column="PRECIOIVA", value=df.PRECIO/1.21) #inserto columna con valor del IVA
```

```
df.head()
```

```
df.round(2) #redondeo la columna que acabo de insertar
```

## **Renombrando columnas**

```
df.columns=df.columns.str.replace("DOCUMENTO","FCNUMERO")
df
```

```
df.columns=df.columns.str.replace("ORGANIZACION","CLIENTE")
df
```

```
df.rename(columns={'IMPORTEMONTRANSACCION':'IMPORTE'}, inplace=True)
df
```

# **Gráficos**

## **Desarrollar gráficos del DataFrame**

```
agruparcliente=df.groupby("CLIENTE").sum()
```

```
px.line(df,
        x="CLIENTE",
        y="CANTIDAD",
        color="PRODUCTO",
        title="Cantidad comprada por cliente",
        template="gridon",
        hover_name="CANTIDAD")
```

```
px.treemap(df,
           values="IMPORTE",
           path=["CLIENTE"],
           color="CANTIDAD",
           hover_name="CLIENTE",
           height=500,
           title="Clientes por ventas")
```

```
alt.Chart(agruparcliente).mark_circle().encode(
  alt.X("PRECIOIVA", bin=True),
  alt.Y("IMPORTE", bin=True),
  color="PRECIOIVA",
  size="PRECIOIVA",
  ).interactive()
```
